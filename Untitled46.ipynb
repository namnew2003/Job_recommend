```python
def cleaningENG(txt):
#     step 1
    txt = re.sub(r'[^a-zA-Z0-9\s]','',txt)
#     step 2
    tokens = nltk.word_tokenize(txt.lower())
    # step 3 and 5
    stemming = [ps.stem(w) for w in tokens if w not in stopwordslisteng]
    return " ".join(stemming)
```

```python
df['Translate_Skills']=df['Translate_Skills'].astype(str).apply(lambda x: cleaningENG(x))
df['Translate_Job_Description']=df['Translate_Job_Description'].astype(str).apply(lambda x: cleaningENG(x))

df['Translate_Career']=df['Translate_Career'].astype(str).apply(lambda x: cleaningENG(x))
```
